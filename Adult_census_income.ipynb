```python
# Cell 4
plt.figure(figsize=(6,4))
sns.countplot(x="income", hue="income", data=df, palette="Set2", legend=False)
plt.title("Target Variable Distribution (Income)")
plt.show()
```

```python

plt.figure(figsize=(12,6)) 
sns.countplot(x="education", hue="education", data=df, palette="Set2", legend=False)
plt.title("Education")

plt.xticks(rotation=45)     
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="marital.status", hue="sex", data=df, palette="coolwarm")
plt.title("Marital Status")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="race", hue="income", data=df, palette="muted")
plt.title("Race vs Income")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x="age", hue="income", data=df, bins=30, palette="Set2")
plt.title("Age Distribution by Income")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x="capital.gain", hue="income", data=df, bins=50, palette="coolwarm", log_scale=(True, False))
plt.title("Capital Gain Distribution by Income (log scale)")
plt.show()
```

```python
#bivariate
plt.figure(figsize=(10,6))
sns.scatterplot(x="age", y="hours.per.week", hue="income", data=df, alpha=0.6, palette="Set2")
plt.title("Age vs Hours per Week by Income")
plt.show()
```

```python
#Categorical vs numerical
plt.figure(figsize=(12,6))
sns.boxplot(x="workclass", y="hours.per.week", hue="income", data=df, palette="pastel")
plt.title("Workclass vs Hours per Week by Income")
plt.xticks(rotation=45)
plt.show()
```

```python
#target varable vs other columns
plt.figure(figsize=(12,6))
sns.countplot(x="workclass", hue="income", data=df, palette="Set2")
plt.title("Workclass vs Income")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="education", hue="income", data=df, palette="Set1")
plt.title("Education vs Income")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="marital.status", hue="income", data=df, palette="muted")
plt.title("Marital Status vs Income")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.countplot(x="occupation", hue="income", data=df, palette="pastel")
plt.title("Occupation vs Income")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="relationship", hue="income", data=df, palette="coolwarm")
plt.title("Relationship vs Income")
plt.show()
```
